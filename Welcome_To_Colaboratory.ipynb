```
#PART 3 standalone function
def data(filename):
    stock_data = pd.read_csv(filename)
    coordinates = list(zip(stock_data['Date'], stock_data['Price']))
    return coordinates
```

```
def moving_average(coordinates, window_size = 20):
    df = pd.DataFrame(coordinates, columns=['Date', 'Price'])
    df['SMA'] = df['Price'].rolling(window=window_size).mean()
    return df
```

```
stock_data = data("stocks.csv")
df_with_moving_average = moving_average(stock_data)

print(df_with_moving_average)
```

```
#PART 4 standalone function
def data(filename):
    stock_data = pd.read_csv(filename)
    coordinates = list(zip(stock_data['Date'], stock_data['Price']))
    return coordinates
```

```
def exponential_moving_average(coordinates, span=20):
    df = pd.DataFrame(coordinates, columns=['Date', 'Price'])
    df['EMA'] = df['Price'].ewm(span=span, adjust=False).mean()
    ema_coordinates = list(zip(df['Date'], df['EMA']))
    return ema_coordinates
```

```
stock_data = data("stocks.csv")
ema_data = exponential_moving_average(stock_data)

for date, ema_price in ema_data:
    print(f"Date: {date}, EMA Price: {ema_price}")
```

```
#PART 5 data obtained in PART 2, 3, and 4 with the same x-coordinate which is the date.
def data(filename):
    stock_data = pd.read_csv(filename)
    stock_data['Date'] = pd.to_datetime(stock_data['Date'])
    return stock_data
```

```
def moving_average(stock_data, window_size = 20):
    stock_data['Moving Average'] = stock_data['Price'].rolling(window=window_size).mean()
    return stock_data
```

```
def exponential_moving_average(stock_data, span = 20):
    stock_data['EMA'] = stock_data['Price'].ewm(span=span, adjust=False).mean()
    return stock_data
```

```
stock_data = data("stocks.csv")

stock_data = moving_average(stock_data)
stock_data = exponential_moving_average(stock_data)
```

```
plt.figure(figsize=(12, 6))
plt.plot(stock_data['Date'], stock_data['Price'], label='Stock Price')
plt.plot(stock_data['Date'], stock_data['Moving Average'], label='20-Day Moving Average')
plt.plot(stock_data['Date'], stock_data['EMA'], label='Exponential Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Time-series forecasting')
plt.legend()
plt.grid(True)
plt.show()
```
